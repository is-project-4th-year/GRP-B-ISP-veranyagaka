```python
import matplotlib.pyplot as plt
from PIL import Image

# Path to one NIH image (change this to your file location)
image_path = "person1_bacteria_1.jpeg"

# Open image
img = Image.open(image_path)

# Show original
plt.imshow(img, cmap="gray")
plt.title("Original NIH X-ray")
plt.axis("off")
plt.show()

```

```python
# Downsample (resize) - e.g. to 128x128
downsampled_img = img.resize((128, 128))

# Show downsampled
plt.imshow(downsampled_img, cmap="gray")
plt.title("Downsampled (128x128)")
plt.axis("off")
plt.show()
```
